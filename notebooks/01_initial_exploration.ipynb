# 01 – Initial Data Exploration
First look at the US Accidents dataset: shape, dtypes, missing values, basic statistics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid', palette='muted')
%matplotlib inline
```

```python
# Load a sample for quick exploration
df = pd.read_csv('../data/raw/US_Accidents_March23.csv', nrows=200_000, low_memory=False)
print(f'Shape: {df.shape}')
df.head(3)
```

```python
print('=== Data Types ===')
print(df.dtypes.value_counts())
print('\n=== Memory Usage ===')
print(f'{df.memory_usage(deep=True).sum() / 1e6:.1f} MB')
```

```python
print('=== Missing Values (%) ===')
missing = (df.isnull().mean() * 100).sort_values(ascending=False)
print(missing[missing > 0].round(2))
```

```python
# Missing value heatmap (sample)
fig, ax = plt.subplots(figsize=(14, 4))
sns.heatmap(df.isnull().sample(500).T, cbar=False, ax=ax, yticklabels=True)
ax.set_title('Missing Value Pattern (500-row sample)')
plt.tight_layout()
plt.show()
```

```python
print('=== Severity Distribution ===')
print(df['Severity'].value_counts())
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df['Severity'].value_counts().sort_index().plot(kind='bar', ax=axes[0], color=sns.color_palette('muted', 4))
axes[0].set_title('Accident Count by Severity')
axes[0].set_xlabel('Severity')
df['Severity'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%', colors=sns.color_palette('muted', 4))
axes[1].set_title('Severity Distribution')
plt.tight_layout()
plt.show()
```

```python
print('=== Numeric Summary ===')
df.describe().T.round(2)
```

```python
print('=== Categorical Columns ===')
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
for col in cat_cols:
    print(f'{col}: {df[col].nunique()} unique values | top: {df[col].value_counts().index[0]}')
```

```python
# Geographic spread
fig, ax = plt.subplots(figsize=(12, 6))
sample = df.dropna(subset=['Start_Lat', 'Start_Lng']).sample(10_000)
ax.scatter(sample['Start_Lng'], sample['Start_Lat'], s=0.5, alpha=0.3, c=sample['Severity'], cmap='RdYlGn_r')
ax.set(title='Accident Locations (10K sample)', xlabel='Longitude', ylabel='Latitude')
plt.colorbar(ax.collections[0], ax=ax, label='Severity')
plt.tight_layout()
plt.show()
```

```python
# Top 15 states
fig, ax = plt.subplots(figsize=(12, 4))
df['State'].value_counts().head(15).plot(kind='bar', ax=ax, color='steelblue')
ax.set_title('Top 15 States by Accident Count')
plt.tight_layout()
plt.show()
```
